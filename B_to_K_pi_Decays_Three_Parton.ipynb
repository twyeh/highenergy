# End-point Divergences
In the collinear factorization approach, one always encounter end-point divergences if the parton distribution amplitudes have no sufficient suppression effects in the end-point region of the collinear momentum configuration.

One typcal example from the annihilation topology has the functional form of $u$
\begin{equation}
X_A=\int_0^1 \frac{\phi_p(u)}{u^n}=\text{div}(\lim_{u\to 0}\frac{u^{x+1-n}}{x+1-n})+\int_0^1\frac{du}{u^n}(\phi_p(u)-\phi_p(0))
\end{equation}
where $\lim_{u\to 0} \phi_p(u)\sim u^x$ with $0\leq x \leq 1$.
The divergent term $\text{div}$ only diver in the double limits $u, x\to 0$. Any limit can only reach a finite result.
