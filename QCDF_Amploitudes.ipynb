```python
import sympy
from sympy import Matrix, init_printing
from sympy.printing.mathml import mathml
from sympy import latex, preview
init_printing()
```

```python
PiZero, PiPlus, PiMinus, KZero, KBarZero, KPlus, KMinus, Eta, EtaPrime, EtaS, EtaSPrime, BBar, B, BPlus, BMinus, BSBar = sympy.symbols([r'\pi^0', r'\pi^+', r'\pi^-', r'K^0', r'\bar{K}^0', r'K^+', r'K^-', r'\eta', r'\eta^{\prime}', r'\eta_s', r'\eta_s^{\prime}', r'\bar{B}', r'B', r'B^+', r'B^-', r'\bar{B}_s'])
```

```python
PseudoMatrix = sympy.Matrix([
    [(PiZero+Eta+EtaPrime)/sympy.sqrt(2), PiMinus, KMinus],  # Example row 1
    [PiPlus, (-PiZero+Eta+EtaPrime)/sympy.sqrt(2), KBarZero],  # Example row 2
    [KPlus, KZero, EtaS+EtaSPrime]   # Example row 3
])

print("Your matrix:")
print(latex(PseudoMatrix))
```

The psudoscalar meson matrix
\begin{equation}
P=\left[\begin{matrix}\frac{\sqrt{2} \left(\eta + \eta^{\prime} + \pi^{0}\right)}{2} & \pi^{-} & K^{-}\\\pi^{+} & \frac{\sqrt{2} \left(\eta + \eta^{\prime} - \pi^{0}\right)}{2} & \bar{K}^0\\K^{+} & K^{0} & \eta_{s} + \eta^{\prime}_{s}\end{matrix}\right]
\end{equation}


```python
RhoZero, RhoPlus, RhoMinus, KSZero, KSBarZero, KSPlus, KSMinus, Omega, Phi, OmegaS, PhiS = sympy.symbols([r'\rho^0', r'\rho^+', r'\rho^-', r'K^{*0}', r'\bar{K}^{*0}', r'K^{*+}', r'K^{*-}', r'\omega', r'\phi', r'\omega_s', r'\phi_s'])
```

```python
VectorMatrix = sympy.Matrix([
    [(RhoZero+Omega+Phi)/sympy.sqrt(2), RhoMinus, KSMinus],  # Example row 1
    [RhoPlus, (-RhoZero+Omega+Phi)/sympy.sqrt(2), KSBarZero],  # Example row 2
    [KSPlus, KSZero, OmegaS+PhiS]   # Example row 3
])

print("Your matrix:")
print(latex(VectorMatrix))
```

\begin{equation}
V=\left[\begin{matrix}\frac{\sqrt{2} \left(\omega + \phi + \rho^{0}\right)}{2} & \rho^{-} & K^{*-}\\\rho^{+} & \frac{\sqrt{2} \left(\omega + \phi - \rho^{0}\right)}{2} & \bar{K}^{*0}\\K^{*+} & K^{*0} & \omega_{s} + \phi_{s}\end{matrix}\right]
\end{equation}

```python
LambdaD = sympy.symbols(r'\lambda^{(d)}_p')
LambdaS = sympy.symbols(r'\lambda^{(s)}_p')
DeltaPU = sympy.symbols(r'\delta_{pu}')
AlphaOne = sympy.symbols(r'\alpha_1')
AlphaTwo = sympy.symbols(r'\alpha_2')
AlphaThree = sympy.symbols(r'\alpha_3^p')
AlphaThreeEW = sympy.symbols(r'\alpha_{3EW}^{p}')
AlphaFour = sympy.symbols(r'\alpha_4^p')
AlphaFourEW = sympy.symbols(r'\alpha_{4EW}^{p}')
```

```python
print(latex(AlphaOne))
print(latex(AlphaTwo))
print(latex(AlphaThree))
print(latex(AlphaThreeEW))
print(latex(AlphaFour))
print(latex(AlphaFourEW))
```

```python
from sympy import Identity
IMatrix=sympy.Matrix([
     [1,0,0],[0,1,0],[0,0,1]
])
BVector=sympy.Matrix([
    [BMinus, BBar, BSBar]
])
LambdaVector=sympy.Matrix([
    [0, LambdaD,LambdaS]
])
UpMatrix=sympy.Matrix([
    [DeltaPU, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
])
QMatrix=sympy.Matrix([
    [1, 0, 0],
    [0, -sympy.Rational(1, 2) , 0],
    [0, 0, -sympy.Rational(1, 2)]
])
print(latex(IMatrix))
print(latex(BVector))
print(latex(LambdaVector.T))
print(latex(UpMatrix))
print(latex(QMatrix))
```

\begin{equation}
I=\left[\begin{matrix}1 & 0 & 0\\0 & 1 & 0\\0 & 0 & 1\end{matrix}\right]\\
B=\left[\begin{matrix}B^{-} & \bar{B} & \bar{B}_s\end{matrix}\right]\\
\Lambda_p=\left[\begin{matrix}0 \\ \lambda^{(d)}_p \\ \lambda^{(s)}_p\end{matrix}\right]\\
U_p=\left[\begin{matrix}\delta_{pu} & 0 & 0\\0 & 0 & 0\\0 & 0 & 0\end{matrix}\right]\\
\hat{Q}=\left[\begin{matrix}1 & 0 & 0\\0 & - \frac{1}{2} & 0\\0 & 0 & - \frac{1}{2}\end{matrix}\right]
\end{equation}

```python
AlphaClassOneMatrix = (AlphaOne * UpMatrix + AlphaFour * IMatrix + AlphaFourEW * QMatrix)
print(latex(AlphaClassOneMatrix))
```

\begin{equation}
\left[\begin{matrix}\alpha_{1} \delta_{pu} + \alpha^{p}_{4} + \alpha_{4EW}^{p} & 0 & 0\\0 & \alpha^{p}_{4} - \frac{\alpha_{4EW}^{p}}{2} & 0\\0 & 0 & \alpha^{p}_{4} - \frac{\alpha_{4EW}^{p}}{2}\end{matrix}\right]
\end{equation}

```python
AlphaClassTwoMatrix = (AlphaTwo * UpMatrix + AlphaThree * IMatrix + AlphaThreeEW * QMatrix)
print(latex(AlphaClassTwoMatrix))
```

\begin{equation}
\left[\begin{matrix}\alpha_{2} \delta_{pu} + \alpha^{p}_{3} + \alpha_{3EW}^{p} & 0 & 0\\0 & \alpha^{p}_{3} - \frac{\alpha_{3EW}^{p}}{2} & 0\\0 & 0 & \alpha^{p}_{3} - \frac{\alpha_{3EW}^{p}}{2}\end{matrix}\right]
\end{equation}

### Decays with $\Delta S=1$ of $B\to PP$

```python
AmplitudeClassOne = BVector * PseudoMatrix * AlphaClassOneMatrix * PseudoMatrix * LambdaVector.T
# AmplitudeClassOne = AmplitudeClassOne.simplify()
print(latex(AmplitudeClassOne))
```

\begin{equation}
\left[\begin{matrix}\lambda^{(d)}_p \left(K^{0} \left(\alpha^{p}_{4} - \frac{\alpha_{4EW}^{p}}{2}\right) \left(B^{-} K^{-} + \bar{B} \bar{K}^0 + \bar{B}_s \left(\eta_{s} + \eta_s^{\prime}\right)\right) + \pi^{-} \left(\alpha_{1} \delta_{pu} + \alpha^{p}_{4} + \alpha_{4EW}^{p}\right) \left(\frac{\sqrt{2} B^{-} \left(\eta + \eta^{\prime} + \pi^{0}\right)}{2} + K^{+} \bar{B}_s + \bar{B} \pi^{+}\right) + \frac{\sqrt{2} \left(\alpha^{p}_{4} - \frac{\alpha_{4EW}^{p}}{2}\right) \left(\eta + \eta^{\prime} - \pi^{0}\right) \left(B^{-} \pi^{-} + K^{0} \bar{B}_s + \frac{\sqrt{2} \bar{B} \left(\eta + \eta^{\prime} - \pi^{0}\right)}{2}\right)}{2}\right) + \lambda^{(s)}_p \left(K^{-} \left(\alpha_{1} \delta_{pu} + \alpha^{p}_{4} + \alpha_{4EW}^{p}\right) \left(\frac{\sqrt{2} B^{-} \left(\eta + \eta^{\prime} + \pi^{0}\right)}{2} + K^{+} \bar{B}_s + \bar{B} \pi^{+}\right) + \bar{K}^0 \left(\alpha^{p}_{4} - \frac{\alpha_{4EW}^{p}}{2}\right) \left(B^{-} \pi^{-} + K^{0} \bar{B}_s + \frac{\sqrt{2} \bar{B} \left(\eta + \eta^{\prime} - \pi^{0}\right)}{2}\right) + \left(\alpha^{p}_{4} - \frac{\alpha_{4EW}^{p}}{2}\right) \left(\eta_{s} + \eta_s^{\prime}\right) \left(B^{-} K^{-} + \bar{B} \bar{K}^0 + \bar{B}_s \left(\eta_{s} + \eta_s^{\prime}\right)\right)\right)\end{matrix}\right]
\end{equation}

```python
from sympy import Trace
AmplitudeClassTwo = BVector * PseudoMatrix * LambdaVector.T * Trace(AlphaClassTwoMatrix * PseudoMatrix)

# AmplitudeClassTwo = AmplitudeClassTwo.simplify()
print(latex(AmplitudeClassTwo))
```
