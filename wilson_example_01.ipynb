```python
from wilson import Wilson
```

```python
wc_dict = {'qq1_1111': 1}
wc = Wilson(wc_dict, scale=91.2, eft='SMEFT', basis='Warsaw')
```

```python
wc_low = wc.match_run(scale=91.2, eft='WET', basis='flavio')
```

```python
wc_bottom = wc.match_run(scale=4.8, eft='WET-3', basis='flavio')
```

```python
wc_bottom
```

```python
wc_bottom_keys=wc_bottom.dict.keys()
```

```python
for key in wc_bottom_keys:
  value=wc_bottom.dict.get(key)
  print(key + " : " + str(value))
```
