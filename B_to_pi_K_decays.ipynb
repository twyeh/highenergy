inspect the experimental data of the particle data group for the Bottom meson decaying into the pion and kaon. collect their branch ratio and asymmetry ratios



*   The pdg library provides an interface to access the PDG data.
We first define the decay process of interest.
* pdg.get_data() fetches the data for the specified decay process.
* We iterate through the data and extract the branching ratio and asymmetry ratio, if available.
* The extracted data is stored in a Pandas DataFrame for easy analysis and presentation.
* Finally, we display the DataFrame to view the results.

```python
import pdg
import pandas as pd
```

```python
# define the API
api = pdg.connect()
for bf in api.get_particle_by_name('B+').exclusive_branching_fractions():
    print('%-60s    %4s    %s' % (bf.description, bf.is_limit, bf.value))
```

```python
for decay in api.get_particle_by_name('B0').exclusive_branching_fractions():
    decay_products = [p.item.name for p in decay.decay_products]
    if 'J/psi(1S)' in decay_products:
        print(format(decay.description,'40s'), decay.display_value_text)
```

```python
# Define the decay process
decay_process = "B+ -> pi+ K0"  # Replace with your desired decay

# Fetch data from PDG
data = pdg.get(decay_process)
```

```python
# Create a list to store the data
results = []

# Loop through the data and extract the desired values
for entry in data:
    try:
        branching_ratio = entry.branching_ratio
        asymmetry_ratio = entry.asymmetry_ratio
        results.append([entry.experiment, branching_ratio, asymmetry_ratio])
    except AttributeError:
        # Skip entries without branching ratio or asymmetry ratio
        pass

# Create a Pandas DataFrame
df = pd.DataFrame(results, columns=["Experiment", "Branching Ratio", "Asymmetry Ratio"])
```

```python
# Print the DataFrame
print(df)
```
