```python
from wilson import Wilson
import pandas as pd
# 1. Initialize the Wilson instance at a high New Physics scale (e.g., 1000 GeV)
# Defining a specific dimension-6 operator (e.g., SMEFT Warsaw basis)
w_smeft = Wilson({'qq1_2323': 1e-6}, scale=1000, eft='SMEFT', basis='Warsaw')

# 2. Automatically run down and match onto the Weak Effective Theory (WET) at the EW scale
wc_wet = w_smeft.match_run(scale=160, eft='WET', basis='flavio')
# To fix the ValueError, create the DataFrame by orienting the dictionary to use keys as index
df = pd.DataFrame.from_dict(wc_wet.dict, orient='index', columns=['value'])
# 3. View the generated low-energy Wilson coefficients
print(df)
```

```python
def query_wilson_operators(wilson_coefficients_dict, filter_string=None):
    """
    Queries and lists operator names from a Wilson coefficients dictionary.

    Args:
        wilson_coefficients_dict (dict): A dictionary of Wilson coefficients,
                                         where keys are operator names.
        filter_string (str, optional): A substring to filter operator names.
                                       Only operators containing this string will be returned.
                                       Defaults to None (return all operators).

    Returns:
        list: A list of operator names.
    """
    operators = list(wilson_coefficients_dict.keys())
    if filter_string:
        operators = [op for op in operators if filter_string.lower() in op.lower()]
    return operators

print("\n--- All Operator Names (first 10) ---")
all_operators = query_wilson_operators(wc_wet.dict)
for op in all_operators[:10]:
    print(op)

print("\n--- Operators containing 'busd' ---")
bsbs_operators = query_wilson_operators(wc_wet.dict, filter_string='busd')
for op in bsbs_operators[:10]:
    print(op)
```
