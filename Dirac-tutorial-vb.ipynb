## 3) Spin Matrices and more
explicit form of $\gamma^\mu$ and $\gamma_5=\gamma^5$:

```python
hep.g3
```

```python
hep.g5
```

$\sigma^{\mu\nu}=\frac{\mathrm i}{2}(\gamma^\mu\gamma^\nu-\gamma^\nu\gamma^\mu)$ here for $\sigma^{ij}$

```python
S1=I/2*(hep.g2*hep.g3-hep.g3*hep.g2);S1
```

```python
S2=I/2*(hep.g3*hep.g1-hep.g1*hep.g3);S2
```

```python
S3=I/2*(hep.g1*hep.g2-hep.g2*hep.g1);S3
```

The $S_i$ form a fourdimensional representation of the Pauli matrices. Proof:

```python
S1*S1+S2*S2+S3*S3
```

```python
I/2*(S1*S3-S3*S1)
```

$S_3/2=\frac{\mathrm i}{2}(\gamma^1\gamma^2-\gamma^2\gamma^1)$ may serve as the spin operator for solutions of the Dirac equation in the rest system. There are four solutions, the normalization is already included.

```python
w1=hep.u(pe0,1);w1
```

```python
w2=hep.u(pe0,1)
```

```python
w3=hep.v(pe0,-1);w3
```

```python
w4=hep.v(pe0,1)
```

$w_1,w_2$ are solutions of positive energy and spins up and down along the z-axis.
$w_3$ and $w_4$ are likewise negative energy solutions. The spin assignment can be checked by equations like the following one:

```python
S3*w3-w3
```

The correct interpretation of the $w_i$ in terms of particles and antiparticles can be read off the rhs of the 
definitions for the $w_i$. Note $w_3$ is an antiparticle with spin down along the z-axis. The operator
$\gamma_5\not\!s$ (where $s$ is the spin four vector) handles the spin assignments properly. In the rest system we
have $s^\mu=(0,0,0,1)$ and $p^\mu=(m,0,0,0)$ with the characteristic invariants $p^2=m^2, s^2=-1,p\cdot s=0$.  The spin operator $H_3$ in the z direction is therefore

```python
H3=-hep.g5*hep.g3
```

```python
H3*hep.v(pe0,1)
```

$\gamma_5\not\!s$  is written invariantly and consequently is also the proper helicity operator in the Lab.
A Lorentz boost transforms the momentum fourvector of the restsystem to $(E,0,0,p)$ and the spin vector to $(p/m,0,0,E/m)$ with $p$ an abbreviation for $\sqrt{E^2-m^2}$. Forgetting the denominator $m$ and the factor 1/2 for convenience we get

```python
H3=hep.g5*hep.g0*sqrt(E**2-m**2)-hep.g5*hep.g3*E;H3
```

```python
H3*hep.v(pez,1)
```

which Jupyter refuses to simplify (see line 17) but which is after dividing by $m$ given by

```python
hep.v(pez,1)
```

which it should be. The textbooks usually quote the helicity projection operator $\Sigma=(1+\gamma_5\not\!s)/2$. Another projection operator is $(1+\gamma_5)/2$ which selects righthandel massless particles

```python
hep.projpl*hep.u(pe,1).subs(m,0)
```

```python
hep.projpl*hep.u(pe,-1).subs(m,0)
```

$\gamma_5$ generically transforms a lefthanded particle into a righthanded antiparticle and vice versa

```python
hep.g5*hep.u(pexyz,-1)
```

```python
hep.v(pexyz,1)
```

## 4) C,P,T

Charge conjugation  $C$ transforms a particle with given helicity into an antiparticle with the same helicity.
$C={\mathrm i}\gamma^2$ with an subsequent complex conjugate transformation does the job:

```python
I*hep.g2*hep.v(pexyz,1)
```

```python
hep.u(pexyz,1)
```

In the xz plane $C={\mathrm i}\gamma^2$

```python
C=I*hep.g2;C
```

```python
C*hep.u(pe,1)
```

```python
hep.v(pe,1)
```

that is $Cu(p_e,1)=v(p_e,1)$. The parity operation $P$ transforms $\vec{p}\rightarrow -\vec{p}$ and 
$\lambda\rightarrow-\lambda$. For spinors this is done by $P=\gamma^0$.

```python
P=hep.g0;P
```

```python
P*hep.v(pe,1)
```

```python
hep.v(peneg,-1)
```

that is $Pv(E,\vec{p},1)=v(E,-\vec{p},-1)$. Time reversal $T$ transforms $\vec{p}\rightarrow -\vec{p}$ leaving $\lambda$ unchanged. In applying $T$ to spinors one has to consider the antiunitarian character of the time reversal operator. Ths means that $T=-\gamma^1\gamma^3$ has to be applied to the complex conjugate spinors (see Bjorken Drell). In the xz plane complex conjugation is not relevant and thus:

```python
T=-hep.g1*hep.g3;T
```

```python
T*hep.v(peneg,-1)
```

```python
hep.v(pe,-1)
```

In summary this means $TPC=\gamma_5$

```python
T*P*C-hep.g5
```

The $S$ matrix is $CPT$ invariant ($CPT$ theorem). Therefore scattering amplitudes of fermions are invariant with respect to replacing particles by antiparticles with reversed helicities.

## 5) Weyl representation

The popular Weyl representation is reached by the unitary transformation

```python
U=(hep.one+hep.g5*hep.g0)/sqrt(2);U
```

```python
Uinv=U.inv();Uinv
```

Basically $\gamma_5$ and $\gamma^0$ are interchanged

```python
g5W=U*hep.g5*Uinv;g5W
```

```python
g0W=U*hep.g0*Uinv;g0W
```

```python
g1W=U*hep.g1*Uinv
```

```python
g2W=U*hep.g2*Uinv
```

```python
g3W=U*hep.g3*Uinv
```

A typical Weyl spinor looks complicated at first sight

```python
uW=U*hep.u(pe,1);uW
```

but with the help of e.g.

```python
simplify(uW[0]**2)
```

we obtain simple expressions for the 4 Weyl helicity spinors

```python
uWR=Matrix([[sqrt(E-p)*cos(theta/2)],[sqrt(E-p)*sin(theta/2)],
             [sqrt(E+p)*cos(theta/2)],[sqrt(E+p)*sin(theta/2)]]);uWR
```

```python
uWL=Matrix([[-sqrt(E+p)*sin(theta/2)],[sqrt(E+p)*cos(theta/2)],
             [-sqrt(E-p)*sin(theta/2)],[sqrt(E-p)*cos(theta/2)]]);uWL
```

```python
vWL=Matrix([[-sqrt(E-p)*cos(theta/2)],[-sqrt(E-p)*sin(theta/2)],
             [sqrt(E+p)*cos(theta/2)],[sqrt(E+p)*sin(theta/2)]]);vWL
```

```python
vWR=Matrix([[sqrt(E+p)*sin(theta/2)],[-sqrt(E+p)*cos(theta/2)],
             [-sqrt(E-p)*sin(theta/2)],[sqrt(E-p)*cos(theta/2)]]);vWR
```

The Weyl representation is particularly attractive for massless particles, because the spinors are reduced
to 2 components only, the Dirac equation is replaced by 2 Weyl equations

```python
uWL.subs(p,E)
```

```python
vWR.subs(p,E)
```
